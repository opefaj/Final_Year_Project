```python
import pandas as pd
file = pd.read_csv("aids_recode.csv")

clust_admin_1 = pd.read_csv("cluster_to_admin_1.csv")
clust_admin_2 = pd.read_csv("cluster_to_admin_2.csv")
clust_admin_3 = pd.read_csv("cluster_to_admin_3.csv")

# Rename columns in preparation for joining.
clust_admin_1 = clust_admin_1.rename(columns = {'DHSCLUST':'HIVCLUST'})
clust_admin_2 = clust_admin_2.rename(columns = {'DHSCLUST':'HIVCLUST'})
clust_admin_3 = clust_admin_3.rename(columns = {'DHSCLUST':'HIVCLUST'})

# Join the dataframes. Only join when HIVCLUST = DHSCLUST
c1 = pd.merge(file, clust_admin_1, how='inner', on='HIVCLUST')
c2 = pd.merge(c1, clust_admin_2, how='inner', on='HIVCLUST')
c3 = pd.merge(c2, clust_admin_3, how='inner', on='HIVCLUST')

# Relevant columns in Aids recode file. 
keep_columns = ['HIVCLUST', 'HIVNUMB', 'HIV03', 'HIV05', 'ID_1', 'ID_2', 'ID_3']
c3 = c3[keep_columns]

file.to_csv("aids_data.csv", index=False)
c3.to_csv("complete_aids_data.csv", index=False)
```

```python
def compute_simple_aggregate(region_name, col_name, data, clust_to_admin_file):
    # Count the number of tests taken for each region. Table: Region, Total taken, Total positive
    admin_data = pd.DataFrame(columns=[region_name,'Total_Participants','Total_Positive_HIV'])
    admin_data[region_name] = list(clust_to_admin_file[col_name].unique())
    
    # Set index to region numbers.   
    aids_data = data
    aids_data.set_index(col_name, inplace=True)
    
    # Set index to region numbers and sort index.
    admin_data.set_index(region_name, inplace=True)
    admin_data.fillna(0, inplace=True)
    admin_data.sort_index(ascending=True, inplace=True)

    for index, row in aids_data.iterrows():
        region = index;
        result = row['HIV03']
        
        #  Increment the number of participants.
        participants = admin_data.loc[region, 'Total_Participants'] + 1
        admin_data.loc[region, 'Total_Participants'] = participants
        
        #  Increment the number of people with positive results.   
        if(result == 1):
            pos = admin_data.loc[region, 'Total_Positive_HIV'] + 1
            admin_data.loc[region, 'Total_Positive_HIV'] = pos
    
    return admin_data
```

```python
admin_1 = compute_simple_aggregate("Admin_1_Region", "ID_1", c3, clust_admin_1)
admin_2 = compute_simple_aggregate("Admin_2_Region", "ID_2", c3, clust_admin_2)
admin_3 = compute_simple_aggregate("Admin_3_Region", "ID_3", c3, clust_admin_3)

print(admin_1.head())
print(admin_2.head())
print(admin_3.head())

admin_1.to_csv("aids_aggregate_1.csv", index=True)
admin_2.to_csv("aids_aggregate_2.csv", index=True)
admin_3.to_csv("aids_aggregate_3.csv", index=True)
```
