```python
# ACTIVITY
activity_admin_1['Activity'] = np.sqrt(activity_admin_1['Activity'])
activity_zscore_1 = stats.zscore(activity_admin_1['Activity'].values)

activity_admin_2['Activity'] = np.sqrt(activity_admin_2['Activity'])
activity_zscore_2 = stats.zscore(activity_admin_2['Activity'].values)

activity_admin_3['Activity'] = np.sqrt(activity_admin_3['Activity'])
activity_zscore_3 = stats.zscore(activity_admin_3['Activity'].values)

# INTROVERSION
introversion_admin_1['avg_introversion'] = np.sqrt(introversion_admin_1['avg_introversion'])
introversion_zscore_1 = stats.zscore(introversion_admin_1['avg_introversion'].values)

introversion_admin_2['avg_introversion'] = np.sqrt(introversion_admin_2['avg_introversion'])
introversion_zscore_2 = stats.zscore(introversion_admin_2['avg_introversion'].values)

introversion_admin_3['avg_introversion'] = np.sqrt(introversion_admin_3['avg_introversion'])
introversion_zscore_3 = stats.zscore(introversion_admin_3['avg_introversion'].values)

# ENTROPY.
entropy_admin_1['normalised_entropy'] = np.sqrt(entropy_admin_1['normalised_entropy'])
entropy_zscore_1 = stats.zscore(entropy_admin_1['normalised_entropy'].values)

entropy_admin_2['normalised_entropy'] = np.sqrt(entropy_admin_2['normalised_entropy'])
entropy_zscore_2 = stats.zscore(entropy_admin_2['normalised_entropy'].values)

entropy_admin_3['normalised_entropy'] = np.sqrt(entropy_admin_3['normalised_entropy'])
entropy_zscore_3 = stats.zscore(entropy_admin_3['normalised_entropy'].values)

# MEDIAN DEGREE.
meddegree_admin_1['medDegree'] = np.sqrt(meddegree_admin_1['medDegree'])
meddegree_zscore_1 = stats.zscore(meddegree_admin_1['medDegree'].values)

meddegree_admin_2['medDegree'] = np.sqrt(meddegree_admin_2['medDegree'])
meddegree_zscore_2 = stats.zscore(meddegree_admin_2['medDegree'].values)

meddegree_admin_3['medDegree'] = np.sqrt(meddegree_admin_3['medDegree'])
meddegree_zscore_3 = stats.zscore(meddegree_admin_3['medDegree'].values)
```
